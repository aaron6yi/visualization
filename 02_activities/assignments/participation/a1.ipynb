```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig,ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.bar(x, y)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.hist(y)  
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(y)  

ax.set_title("total groth over time")
ax.set_xlabel("year since staqrt")
ax.set_ylabel("total growth")
```

```python
font1 ={'family' : 'sans-serif','color':'blue','size':20}
font2 ={'family' : 'monospace','color':'red','size':30}
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(y)  
ax.set_title("total groth over time",fontdict = font1)
ax.set_xlabel("year since staqrt",fontdict = font2)
ax.set_ylabel("total growth",fontdict = font1)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, linestyle='--', linewidth=2, color='indigo') 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker ='*', linestyle='--', linewidth=2, color='indigo') 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker ='*', linestyle='--', linewidth=2, color='indigo', markeredgecolor = '#FA9359', markerfacecolor = '#000000') 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker ='*', linestyle='--', linewidth=2, color='indigo', markeredgecolor = '#FA9359', markerfacecolor = '#000000') 
ax.grid('y')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker ='*', linestyle='--', linewidth=2, color='indigo', markeredgecolor = '#FA9359', markerfacecolor = '#000000') 
ax.grid(axis='y', color='gray', linestyle='-.', linewidth=1)
plt.show()
```

```python
sns.jointplot(x=x, y=y, kind='scatter', color='indigo')
plt.show()

sns.jointplot(x=x, y=y, kind='hex', color='purple')
plt.show()

sns.jointplot(x=x, y=y, kind='kde', fill=True, color='blue')
plt.show()
```
