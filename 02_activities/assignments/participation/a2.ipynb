```python
# Importing the NumPy library for numerical operations and handling arrays
import numpy as np

# Importing Matplotlib for data visualization
import matplotlib.pyplot as plt

# Importing Seaborn for statistical data visualization
import seaborn as sns

# Importing Pandas for data manipulation and analysis
import pandas as pd

# Importing SciPy for scientific computations, such as optimization, statistics, and signal processing
import scipy

```

```python
# Setting the random seed for reproducibility
np.random.seed(613)

# Creating an array of 50 sequential integers from 0 to 49
x = np.arange(50)

# Generating 50 random integers between 0 and 100
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Creating a figure and axis for the plot, with a specified figure size of 5x3 inches
fig, ax = plt.subplots(figsize=(5, 3))

# Plotting the data in y as a line plot
# The x-axis automatically represents the index positions (0 to len(y)-1)
ax.plot(x,y1, label="person1")
ax.plot(x,y2, label="person2")
ax.legend(loc = "lower right",
          frameon=True,
          fontsize="x-large",
          ncol = 2,
          shadow =True)
# Adding a title to the plot
ax.set_title("Total Growth Over Time")

# Labeling the x-axis
ax.set_xlabel("Years Since Start")

# Labeling the y-axis
ax.set_ylabel("Total Growth")

```
