```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("fivethirtyeight")

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))

font_title = {'family': 'serif', 'color': 'darkblue', 'size': 16}
font_labels = {'family': 'serif', 'color': 'black', 'size': 12}

ax1.scatter(x1, y1, color='blue', label='Random Data')
ax1.set_title('Random Integer Scatter Plot', fontdict=font_title)
ax1.set_xlabel('Index', fontdict=font_labels)
ax1.set_ylabel('Random Values', fontdict=font_labels)
ax1.legend(loc='upper right', fontsize=10)

max_index = np.argmax(y1)
ax1.annotate(
    'Max Value', xy=(x1[max_index], y1[max_index]), xytext=(30, 70),
    arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12
)
ax1.text(10, 10, 'Random Points', fontsize=12, color='purple')

ax2.bar(x2, y2, color='green', label='Characters Scores')
ax2.set_title('Character Scores', fontdict=font_title)
ax2.set_xlabel('Characters', fontdict=font_labels)
ax2.set_ylabel('Scores', fontdict=font_labels)
ax2.legend(loc='upper right', fontsize=10)

max_score_index = np.argmax(y2)
ax2.annotate(
    'Highest Score', xy=(x2[max_score_index], y2[max_score_index]), xytext=(2, 250),
    arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=12
)
ax2.text(0, 50, 'Character Analysis', fontsize=12, color='darkred')

plt.tight_layout()

plt.show()
```
